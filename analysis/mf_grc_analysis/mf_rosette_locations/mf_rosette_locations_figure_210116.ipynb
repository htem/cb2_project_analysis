```python
n_mfs = 0
for mf_id, mf in input_graph.mfs.items():
    do_count = False
    for loc in mf.locs:
        x, y, z = loc
        if x < x_min*1000 or x > x_max*1000:
            continue
        if z < z_min*1000 or z > z_max*1000:
            continue
        do_count = True
    if do_count:
        n_mfs += 1
print(f'n_mfs = {n_mfs} ({z_min}-{z_max})')
```
